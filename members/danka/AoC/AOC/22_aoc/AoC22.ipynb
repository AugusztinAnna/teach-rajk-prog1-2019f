```python
def newstack(mydeck):
    newstack = mydeck[::-1]
    return newstack
```

```python
def cut(mydeck, n):
    newstack = mydeck[n:] + mydeck[:n]
    return newstack
```

```python
def increment(mydeck, n):
    i = 0
    newdeck = list(range(0, len(mydeck)))
    for c in range(len(mydeck)):
        if i > len(mydeck):
            i = i % len(mydeck)
            newdeck[i] = mydeck[c]
            i += n
        else:
            newdeck[i] = mydeck[c]
            i += n
    return newdeck
```

```python
mydeck = list(range(0, 10))
increment(mydeck, 3)
```

```python
def aoc22(input_list):
    mydeck = list(range(0, 10007))
    for shuffle in input_list:
        if shuffle[:19] == 'deal with increment':
            n = int(shuffle[-2:])
            mydeck = increment(mydeck, n)
            print(n)
        if shuffle[:3] == 'cut':
            n = int(shuffle[3:])
            mydeck = cut(mydeck, n)
            print(n)
        if shuffle[:19] == 'deal into new stack':
            mydeck = newstack(mydeck)
            print("newstack")
            
    return mydeck
```
