```python
with open("input7.txt", "r", encoding="utf-8") as fp:
    raw_input = fp.read().split(",")
raw_input
input_list = []
for strings in raw_input:
    input_list.append(int(strings))
print(len(input_list))
```

```python
from itertools import permutations

permut_list = list(permutations(range(5), 5))

phase_list = []

for elements in permut_list:
    phase_list.append(list(elements))

```

```python
output_signals = []
for phase in phase_list:
    new_input = 0
    for setting in phase:
    
        init_list = input_list.copy()
        start = 0
        new_list = init_list
        output_list = []
        
        count3 = 0
        opc3input = [setting, new_input]

        while start < len(init_list):

            #opcode jó formába
            if len(str(init_list[start])) == 5:
                string = str(init_list[start])
            if len(str(init_list[start])) == 4:
                string = '0' + str(init_list[start])
            if len(str(init_list[start])) == 3:
                string = '00' + str(init_list[start]) 
            if len(str(init_list[start])) == 2:
                string = '000' + str(init_list[start])    
            if len(str(init_list[start])) == 1:
                string = '0000' + str(init_list[start])

            # összeadás
            if string[4] == "1":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    new_list[init_list[start + 3]] = par1 + par2

                if string[0] == 1:
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    init_list[start + 3] = par1 + par2
                start += 4
            # szorzás
            if string[4] == "2":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    new_list[init_list[start + 3]] = par1 * par2
                if string[0] == "1":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    init_list[start + 3] = par1 * par2
                start += 4
            # input
            if string[4] == "3":
                new_list[init_list[start + 1]] = opc3input[count3]
                print("opc3", opc3input[count3])
                count3 += 1
                start += 2
            # output
            if string[4] == "4":
                print(" opc 4 output:", new_list[init_list[start + 1]])
                output_list.append(new_list[init_list[start + 1]])
                start += 2
            # jump-if-true
            if string[4] == "5":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 != 0:
                        start = par2
                    else:
                        start += 3
                if string[0] == "1":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 != 0:
                        start = par2
                    else:
                        start += 3

            # jump-if-false
            if string[4] == "6":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 == 0:
                        start = par2
                    else:
                        start += 3
                if string[0] == "1":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 == 0:
                        start = par2
                    else:
                        start += 3


            # less than
            if string[4] == "7":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 < par2:
                        new_list[init_list[start + 3]] = 1
                    else:
                        new_list[init_list[start + 3]] = 0
                if string[0] == "1":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 < par2:
                        new_list[init_list[start + 3]] = 1
                    else:
                        new_list[init_list[start + 3]] = 0
                start += 4

            # equals
            if string[4] == "8":
                if string[0] == "0":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 == par2:
                        new_list[init_list[start + 3]] = 1
                    else:
                        new_list[init_list[start + 3]] = 0
                if string[0] == "1":
                    if string[1] == "1":
                        par2 = int(init_list[start + 2])
                    else:
                        par2 = int(init_list[init_list[start + 2]])
                    if string[2] == "1":
                        par1 = int(init_list[start + 1])
                    else:
                        par1 = int(init_list[init_list[start + 1]])

                    if par1 == par2:
                        new_list[init_list[start + 3]] = 1
                    else:
                        new_list[init_list[start + 3]] = 0
                start += 4

            # halting
            if string[3:] == "99":
                print("halting...")
                break
            print("start:", start)
        print(output_list)
        new_input = output_list[0]
    output_signals.append(new_input)
```
