```python
with open("input6.txt", "r", encoding="utf-8") as fp:
    raw_input = fp.read().split("\n")

orbit_list = []
for string in raw_input:
    split = []
    split = (string.split(")"))
    orbit_list = orbit_list + split

final_list = [inputlist[x:x+2] for x in range(0, len(inputlist),2)]
```

```python
# YOU

counter = 0
youlist = []

current = ['YOU']

for i in range(len(orbitlist)):
    new_list = []
    for pair in final_list:
        if pair[1] in current:
            print("pair", pair[1])
            new_list.append(pair[0])
            youlist.append(pair[0])
            counter += 1
            current = new_list
print(youlist)



            
```

```python
# SANTA

counter = 0
sanlist = []

current = ['SAN']

for i in range(len(orbitlist)):
    new_list = []
    for pair in final_list:
        if pair[1] in current:
            print("pair", pair[1])
            new_list.append(pair[0])
            sanlist.append(pair[0])
            counter += 1
            current = new_list
            
print(sanlist)
```

```python
sol2 = []
for objects in youlist:
    if objects not in sanlist:
        sol2.append(objects)
        
for objectss in sanlist:
    if objectss not in youlist:
        sol2.append(objectss)

print(len(set(sol2)))
```
