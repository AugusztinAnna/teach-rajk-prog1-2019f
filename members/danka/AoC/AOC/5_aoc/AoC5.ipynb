```python
def aoc5(init_list):
    start = 0
    new_list = init_list
    output_list = []

    while start < len(init_list):
        
        #opcode jó formába
        if len(str(init_list[start])) == 5:
            string = str(init_list[start])
        if len(str(init_list[start])) == 4:
            string = '0' + str(init_list[start])
        if len(str(init_list[start])) == 3:
            string = '00' + str(init_list[start]) 
        if len(str(init_list[start])) == 2:
            string = '000' + str(init_list[start])    
        if len(str(init_list[start])) == 1:
            string = '0000' + str(init_list[start])
                
        # összeadás
        if string[4] == "1":
            if string[0] == "0":
                if string[1] == "1":
                    print("par2 immediate")
                    par2 = int(init_list[start + 2])
                else:
                    print("par2 position, index:", init_list[start + 2])
                    par2 = int(init_list[init_list[start + 2]])
                if string[2] == "1":
                    print("par1 immediate")
                    par1 = int(init_list[start + 1])
                else:
                    print("par1 position, index:", init_list[start + 1])
                    par1 = int(init_list[init_list[start + 1]])
                
                new_list[init_list[start + 3]] = par1 + par2
                
                print("opc1: addition")
            if string[0] == 1:
                if string[1] == "1":
                    print("par2 immediate")
                    par2 = int(init_list[start + 2])
                else:
                    print("par2 position, index:", init_list[start + 2])
                    par2 = int(init_list[init_list[start + 2]])
                if string[2] == "1":
                    print("par1 immediate")
                    par1 = int(init_list[start + 1])
                else:
                    print("par1 position, index:", init_list[start + 1])
                    par1 = int(init_list[init_list[start + 1]])
                
                init_list[start + 3] = par1 + par2
                print("opc1: addition")
            start += 4
        # szorzás
        if string[4] == "2":
            if string[0] == "0":
                if string[1] == "1":
                    print("par2 immediate")
                    par2 = int(init_list[start + 2])
                else:
                    print("par2 position, index:", init_list[start + 2])
                    par2 = int(init_list[init_list[start + 2]])
                if string[2] == "1":
                    print("par1 immediate")
                    par1 = int(init_list[start + 1])
                else:
                    print("par1 position, index:", init_list[start + 1])
                    par1 = int(init_list[init_list[start + 1]])
                    
                new_list[init_list[start + 3]] = par1 * par2
                print("opc2: multiplicaton")
            if string[0] == "1":
                if string[1] == "1":
                    print("par2 immediate")
                    par2 = int(init_list[start + 2])
                else:
                    print("par2 position, index:", init_list[start + 2])
                    par2 = int(init_list[init_list[start + 2]])
                if string[2] == "1":
                    print("par1 immediate")
                    par1 = int(init_list[start + 1])
                else:
                    print("par1 position, index:", init_list[start + 1])
                    par1 = int(init_list[init_list[start + 1]])
                
                init_list[start + 3] = par1 * par2
                print("opc2: multiplication")
            start += 4
        # input
        if string[4] == "3":
            new_list[init_list[start + 1]] = input()
            print("opc3")
            start += 2
        # output
        if string[4] == "4":
            print("output:", new_list[init_list[start + 1]])
            output_list.append(new_list[init_list[start + 1]])
            print("opc4")
            start += 2
        # halting
        if string[3:] == "99":
            print("halting...")
            break
        print("start:", start)
    print(output_list)
    return
```
