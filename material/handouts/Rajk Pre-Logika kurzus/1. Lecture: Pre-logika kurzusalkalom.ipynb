<h2><span id="slide-prelog:muvs-exs">Slide 6: Példák</span></h2>


<h3><span id="content-logic:stmnt_examples">Kijelentések:</span></h3>


A: Dávid okos, B: Dávid tudja ki 2008 aranylabdása, C: Dávid szereti a fagyit, D: kevesen tudják ki 2008 aranylabdása


<h3><span id="content-logic:exformulas">Formulává alakítható</span></h3>


1. Dávid okos és tudja ki 2008 aranylabdása
    - A $\land$ B <!-- .element: class="fragment" -->
1.  Dávid nem szereti a fagyit de okos
    - $\neg$ C $\land$ A <!-- .element: class="fragment" -->
1.  Akkor tudják kevesen ki 2008 aranylabdása, ha Dávid tudja és okos
    - D $\Leftarrow$ (B $\land$ A) <!-- .element: class="fragment" -->
1.  Dávid akkor és csakis akkor okos, ha tudja ki 2008 aranylabdása és ezt kevesen tudják
    - A $\Leftrightarrow$ (B $\land$ D) <!-- .element: class="fragment" -->
1.  Ha kevesen tudják ki 2008 aranylabdása, Dávid nem tudja
    - D $\Rightarrow$ $\neg$ B <!-- .element: class="fragment" -->

<h2><span id="slide-prelog:tauts">Slide 10: Tételek</span></h2>


<h3><span id="content-logic:tautdef">Tétel</span></h3>



olyan formulák amik a bennük szereplő változók értékétől függetlenül igazak



<h3><span id="content-logic:tautlist">Tételek</span></h3>



- asszociativitás:
    - $A \lor (B \lor C) \Leftrightarrow (A \lor B) \lor C$
- disztributivitás:
    - $A \land (B \lor C) \Leftrightarrow (A \land B) \lor (A \land C)$
- De Morgan szabály:
    - $\neg (A \lor B) \Leftrightarrow \neg A \land \neg B$
- szillogizmus:
    - $(A \Rightarrow B) \land (B \Rightarrow C) \Rightarrow (A \Rightarrow C)$

<h2><span id="slide-prelog:predics">Slide 13: Predikátumok</span></h2>


<h3><span id="content-logic:predicexs1">Predikátumok</span></h3>



1. egyváltozós
    - $P$: "páros szám"
    - $P(x)$ akkor igaz ha $x$ páros
2. többváltozós
    - $NE$: "nagyobb egyenlő"
    - $NE(x,y)$ akkor igaz ha $x \geq y$, különben hamis
3. formulában
    - $NE(x,y) \land P(y)$
    - $NE(x,y) \Rightarrow NE(y,x)$



<h2><span id="slide-prelog:quantis">Slide 14: Kvantorok</span></h2>


<h3><span id="content-logic:quantexs">Kvantorok</span></h3>



1.  $\forall$
1.  $\exists$

Formulában:
1.  $\forall x ( P(2x) )$
1.  $\forall x ( \exists y (NE(x,y)))$
1.  $\nexists x (\neg NE(x,x))$



<h2><span id="slide-prelog:formulas">Slide 15: Nyitott és zárt formulák</span></h2>


<h3><span id="content-logic:boundvardef">Kötött paraméter</span></h3>


Ha egy kijelentésben egy paraméter minden előfordulása valamilyen kvantor hatáskörében van, akkor azt mondjuk, hogy a paraméter **kötött**, különben **szabad**


<h3><span id="content-logic:closeformdef">Zárt formula</span></h3>


Az a formula, aminek nincs szabad paramétere, **zárt formula**. Máskülönben **nyitott formula**
